```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
IMAGE_PATH = r"C:\Users\drang\Pictures\Screenshots\Screenshot (21).png"
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(IMAGE_PATH,paragraph="True")
result
```

```python
#img = cv2.imread(IMAGE_PATH)
from skimage import io

image = io.imread(IMAGE_PATH)
img = np.asarray(image)
spacer = 100
font = cv2.FONT_HERSHEY_SIMPLEX
for detection in result: 
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(10,spacer), font, 0.5,(0,100,0),1,cv2.LINE_AA)
    spacer+=15
    with open('readme3.txt', 'a') as f:
        f.write(text)
        f.close()
fig = plt.figure(figsize=(10,10))
#print(img.size)
plt.imshow(img)
plt.savefig("boxtext.png")
plt.show()
plt.close()
```
